```python
import pandas as pd
```

```python
df = pd.DataFrame([
    [0, 17], [0, 13],
    [1, 17], [1, 15], [1, 14], [1, 13],
    [2, 17], [2, 16], [2, 13],
    [3, 13], [3, 12], [3, 11],
    [4, 13], [4, 12], [4, 11], [4,8],
    [5,  9], [5,  8], [5,  7], [5, 6],
    [6,  8], [6,  7], [6,  3],
    [7,  5], [7,  4],
    [8,  1], [8,  2],
    [9,  1],
   [10,  1],
],
columns=['Teaching', 'Flocksize'])
scatter = df.plot.scatter(x='Teaching',
                          y='Flocksize',
                          c='DarkBlue')
```

```python
import matplotlib as mpl
mpl.use('svg')
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
  return 15.5 / (1 + np.exp((z*0.7)-3.5))

# plot sigmoid function for logistic regression
t = np.arange(0., 10., 0.5)
plt.plot(t, sigmoid(t))


plt.ylabel('Max Flock Size')
plt.xlabel('Teaching Interval')
plt.scatter(x='Teaching', y='Flocksize', c='DarkBlue', data = df)

plt.axvline(x=1)
plt.text(0.5,1,'Oblivion Half Life',rotation=90)

plt.show()
plt.savefig("FlockSize.svg", format="svg")
```

```python
def exponential_decay(t):
  return np.exp((-np.log(2)*t))

t = np.arange(0., 10., 0.1)
plt.plot(t, exponential_decay(t))

plt.ylabel('Agent Memory')
plt.xlabel('Simulation Time')
plt.axvline(x=1)
plt.text(0.6,0,'Oblivion Half Life',rotation=90)

plt.show()
plt.savefig("ExponentialDecay.svg", format="svg")
```

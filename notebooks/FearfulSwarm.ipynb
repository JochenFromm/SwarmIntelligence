# Fearful Swarm Model
Evaluation of experiments for fearful swarm model

```python
import pandas as pd
```

```python
df = pd.DataFrame([
    [0, 1],
    [0.1, 1],  [0.1, 2],
    [0.2, 5],  [0.2, 7],  [0.2, 9],  [0.2, 10],
    [0.3, 10], [0.3, 13], [0.3, 16], [0.3, 17],
    [0.4, 10], [0.4, 15], [0.4, 16], [0.4, 18],
    [0.5, 14], [0.5, 15], [0.5, 18],
    [0.6, 13], [0.6, 14], [0.6, 15], [0.6, 17], [0.6, 20],
    [0.7, 15], [0.7, 17], [0.7, 20],
    [0.8, 17], [0.8, 19], [0.8, 20],
    [0.9, 18], [0.9, 19],
    [1.0, 20],
],
columns=['Fearfulness', 'Flocksize'])
scatter = df.plot.scatter(x='Fearfulness',
                          y='Flocksize',
                          c='DarkBlue')
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np

def sigmoid(z):
  return 20 / (1 + np.exp((-z*8.5+2.5)))

# plot sigmoid function for logistic regression
t = np.arange(-0.05, 1.1, 0.05)

fig, ax = plt.subplots()
plt.plot(t, sigmoid(t))
ax.xaxis.set_major_formatter(mtick.PercentFormatter(1,0))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter("%d"))
plt.ylabel('Max Flock Size')
plt.xlabel('Agent Fearfulness')
plt.scatter(x='Fearfulness', y='Flocksize', c='DarkBlue', data = df)
plt.show()
```

```python
mpl.use('svg')
fig, ax = plt.subplots()
plt.plot(t, sigmoid(t))
ax.xaxis.set_major_formatter(mtick.PercentFormatter(1,0))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter("%d"))
plt.ylabel('Max Flock Size')
plt.xlabel('Agent Fearfulness')
plt.scatter(x='Fearfulness', y='Flocksize', c='DarkBlue', data = df)
plt.savefig("FlockSizeFearfulSwarm.svg", format="svg")
```
